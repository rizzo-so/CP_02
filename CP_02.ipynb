# Desafio: Cálculo da média FIAP

```python
# Cálculo de média dos checkpoints

# 3 CPs obrigatórios ---> input
CP1, CP2, CP3 = input('Insira as notas dos checkpoints: ').split()
CP1 = float(CP1)
CP2 = float(CP2)
CP3 = float(CP3)

# Eliminar a menor nota
menor = CP1
if (CP2 < menor):
  menor = CP2
if (CP3 < menor):
  menor = CP3

# Cálcular média das duas maiores notas (MCP_1SEM)
SOMA_CP = (CP1 + CP2 + CP3) - menor
MCP_1SEM = SOMA_CP / 2
print(f'Média das CPs: {MCP_1SEM} ')
```

```python
# Cálculo de notas de SPRITs (SP1 e SP2)
SP1, SP2 = input('Insira as notas das SPRINTs: ').split()
SP1 = float(SP1)
SP2 = float(SP2)

# Média das duas notas
MSP1 = (SP1 + SP2) / 2
print(f'Média das SPRINTs: {MSP1}')
```

```python
# Cálculo de notas Global Solution (GS)

# Inserir nota
GS1 = input('Insira sua nota da GS (Caso tenha feito SUB, deixar vazio): ')

# Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB
if (GS1 != ""):
    GS1 = float(GS1)
    print(f"Nota GS1 = {GS1}")
else:
    GS1 = float(input("GS não realizada. Insira a nota da SUB: "))
    print(f"Nota SUB = {GS1}")
```

```python
# Cálculo da média semestral 1 (MS1)
# MS1 = 0.6*GS1 + 0.2*MSP1 + 0.2*MCP1
MS1 = 0.6*GS1 + 0.2*MSP1 + 0.2*MCP_1SEM
print(f'Média 1 semestre:{MS1:.2f}')
```

```python
# Repetir para o segundo semestre

# Cálculo de média dos checkpoints
# 3 CPs obrigatórios ---> input
CP1_2, CP2_2, CP3_2 = input('Insira as notas dos checkpoints: ').split()
CP1_2 = float(CP1_2)
CP2_2 = float(CP2_2)
CP3_2 = float(CP3_2)

# Eliminar a menor nota
menor = CP1_2
if (CP2_2 < menor):
  menor = CP2_2
if (CP3_2 < menor):
  menor = CP3_2

# Cálcular média das duas maiores notas (MCP_1SEM)
SOMA_CP_2 = (CP1_2 + CP2_2 + CP3_2) - menor
MCP_2SEM = SOMA_CP_2 / 2
print(f'Média das CPs: {MCP_2SEM} ')



# Cálculo de notas de SPRITs (SP1 e SP2)
SP1_2, SP2_2 = input('Insira as notas das SPRINTs: ').split()
SP1_2 = float(SP1_2)
SP2_2 = float(SP2_2)

# Média das duas notas
MSP2 = (SP1_2 + SP2_2) / 2
print(f'Média das SPRINTs: {MSP2}')



# Cálculo de notas Global Solution (GS)
# Inserir nota
GS2 = input('Insira sua nota da GS (Caso tenha feito SUB, deixar vazio): ')

# Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB
if (GS2 != ""):
    GS2 = float(GS2)
    print(f"Nota GS2 = {GS2}")
else:
    GS2 = float(input("GS não realizada. Insira a nota da SUB: "))
    print(f"Nota SUB = {GS2}")



# Cálculo da média semestral 2 (MS2)
# MS1 = 0.6*GS2 + 0.2*MSP2 + 0.2*MCP2
MS2 = 0.6*GS2 + 0.2*MSP2 + 0.2*MCP_2SEM
print(f'Média 1 semestre:{MS2:.2f}')
```

```python
# Média final FIAP
MF = 0.4*MS1 + 0.6*MS2
print(f'Média final FIAP: {MF:.2f}')
```

```python
# Apresentar se o aluno está aprovado, reprovado ou de exame
# Aprovado >= 6.0
# Exame >= 4.0 e < 6.0
# Reprovado < 4.0
if MF >= 6.0:
    print(f"Média Final: {MF:.1f}")
    print("Status: Aprovado!")

else:
    if MF >= 4.0:
        print(f"Média Final: {MF:.1f}")
        print("Status: Exame. Você precisa fazer a prova final.")

        # Se estiver de exame, inserir a nota do exame
        EX = float(input("Digite a nota do Exame: "))

        # Cálcular MF2 = (EX + MF) / 2
        MF2 = (EX + MF) / 2
        print(f"Sua nova média após exame (MF2) é: {MF2:.1f}")

        if MF2 >= 6.0:
            print("Status: Aprovado no Exame!")
        else:
            print("Status: Reprovado após o Exame.")

    else:
        print(f"Média Final: {MF:.1f}")
        print("Status: Reprovado")
```
